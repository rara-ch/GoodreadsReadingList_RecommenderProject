## Exploratory Data Analysis

The only categorical column is the author column.

```python
# Find the average number of books an author has written
books['author'].value_counts().mean()
```

NOTE TO SELF:

```python
# Check the distributions (histogram) of the numeric columns
for column in books.iloc[:, 5:].drop(['description', 'image_url'], axis=1):
    books[column].plot.hist(bins=100)
    plt.title(column)
    plt.show()
```

```python
# Check the distributions (boxplot) of the numeric columns
for column in books.iloc[:, 5:].drop(['description', 'image_url'], axis=1):
    books[column].plot.box()
    plt.title(column)
    plt.show()
```
